```python
# Cuando hay datos faltantes se rellena con la mediana o promedio
valores_por_defecto={"nombre":"Desconocido",
                     "edad": 32.5,
                     "ingreso": 17200}
print(datos.fillna(value=valores_por_defecto))
```

```python
promedio = datos["ingreso"].mean()
mediana = datos["ingreso"].median()
moda = datos["ingreso"].mode()[0]
print(promedio, mediana, moda)
```

```python
promedio = datos["edad"].mean()
mediana = datos["edad"].median()
moda = datos["edad"].mode()[0]
print(promedio, mediana, moda)
```
