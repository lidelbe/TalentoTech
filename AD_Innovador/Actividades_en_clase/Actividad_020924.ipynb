```python
# Actividad en clase: Tabla de pesos y estaturas

# importar librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# crear una base de datos
# Construir la lista de valores

value_n_list = ["Juan", "Maria", "Carlos", "Ximena", "Luis", "Lorena", "Pablo", "Edgar", "Marta", "Javier", "Rosa"]
value_p_list = [70, 65, 80, 55, 75, 62, 85, 58, 68, 60, 75] # [Kg]
value_e_list = [1.70, 1.75, 1.60, 1.80, 1.65, 1.70, 1.68, 1.85, 1.62, 1.75, 1.63] # [mts]
```

```python
# Usar un diccionario ({key:value})
# Con formato {"2str":list}
data_dict = {
    "nombre": value_n_list,
    "peso": value_p_list,
    "estatura": value_e_list
}
```

```python
# crear un DataFrame a partir del diccionario
df = pd.DataFrame(data_dict)
```

```python
# mostrar el DataFrame
print(df)
```

```python
# Visualizar la tabla de pesos y estaturas
plt.figure(figsize=(10,6))
plt.scatter(df['estatura'], df['peso'], color='blue')
plt.title('Gr√°fico de Pesos vs Estaturas')
plt.xlabel('Estatura (m)')
plt.ylabel('Peso (kg)')
plt.grid(True)
plt.show()
```
